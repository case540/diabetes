```python
from itertools import chain
from datasets import load_from_disk
import pickle

train_datsset_path = "/share/home/liangzhongming/930/CGMformer/data/8_11_data/downstream/CV_5/downsampled_CV_5_train_96"
train_dataset=load_from_disk(train_datsset_path)

# test_datsset_path = "/share/home/liangzhongming/930/CGMformer/data/8_11_data/downstream/CV_5/downsampled_CV_5_test_96"
# test_dataset=load_from_disk(train_datsset_path)
```

```python
input_ids1 = train_dataset['input_ids']
all_input_ids = list(chain(input_ids1))

unique_tokens = set()

for sequence in all_input_ids:
    for token in sequence:
        integer_token = int(token)
        unique_tokens.add(integer_token)
unique_tokens

sorted_tokens = sorted(list(unique_tokens))

token2id = {token: index for index, token in enumerate(sorted_tokens)}
token2id["<mask>"] = max(token2id.values()) + 1
token2id["<pad>"] = max(token2id.values()) + 1
token2id["<cls>"] = max(token2id.values()) + 1

id2token = {index: token for token, index in token2id.items()}
```

```python
with open('/share/home/liangzhongming/930/CGMformer/data/8_11_data/downstream/CV_5/CV5_total_token2id.pkl', 'wb') as f:
    pickle.dump(token2id, f)
with open('/share/home/liangzhongming/930/CGMformer/data/8_11_data/downstream/CV_5/CV5_total_id2token.pkl', 'wb') as f:
    pickle.dump(id2token, f)
```
